## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

```python
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
%matplotlib inline

x_rv = np.random.randn(200000)
a = 1.
n_cus = np.random.randn(200000)

def rxy_calc(x,n,a):
    n_rv = n*n_cus
    y_rv = a*x_rv+n_rv
    sxy = np.sum((x_rv-np.mean(x_rv))*(y_rv-np.mean(y_rv)))/199999
    rxy = sxy/(np.std(x_rv)*np.std(y_rv))
    return rxy

n_list = np.logspace(-2,2,40)
rxy_array = np.zeros(len(n_list))

for i in range (0,len(n_list)):
    rxy_array[i] = rxy_calc(x_rv,n_list[i],a)
    
print rxy_array

inv_n_list = np.abs(np.mean(x_rv))*a/n_list #abs value added since plotting in log-log

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(2,1,1)
ax.loglog(inv_n_list,rxy_array)
ax.set_xlabel('a*S_x/n [A.U.]')
ax.set_ylabel('Correlation')
ax.set_title('Correlation decreases with increasing noise amplitude')
```

This plot says that as the noise increases, the random variables x and y become less linearly correlated because the noise term is becoming more dominant than the linear term. It looks like the functional form is Lorenzian-esque, having a corner frequency at about 4e-4.

**2** Repeat the above, but with a different value of $a$.  Comment on how (and why) $a$ changes the value of $r_{xy}$.  

```python
b = 4.

rxy_array_b = np.zeros(len(n_list))

for i in range (0,len(n_list)):
    rxy_array_b[i] = rxy_calc(x_rv,n_list[i],b)
    
inv_n_list_2 = np.abs(np.mean(x_rv))*b/n_list 

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(2,1,1)
ax.loglog(inv_n_list,rxy_array,label='a=1')
ax.loglog(inv_n_list_2,rxy_array_b,label='b=4')
ax.set_xlabel('a*S_x/n [A.U.]')
ax.set_ylabel('Correlation')
ax.set_title('Correlation decreases with increasing noise amplitude')
ax.legend()

print np.mean(rxy_array), np.mean(rxy_array_b)
```

The second random variable has a larger linear coeffecient and so the linear correlation is less affected by the noise, which is seen in the figure by noticing that mean(signal2) is higher than mean(signal1). The graphs superimpose because the x axis has been normalized by a. Interestingly, the decay of the correlation appears to follow the same power law for linear coeffecients a *and* b.

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

```python
n_num = 1. ; a = 1.
data_num = 10

def slope_finder(data_num, n_num, a):
    x = np.random.randn(data_num)
    n = n_num*np.random.randn(data_num)
    y = a*x+n
    y_til = a*x
    
    slope, intercept = np.polyfit(x, y, 1)
    syx = np.sqrt(np.sum((y-y_til)**2/(data_num-2)))
    factor = np.sqrt(np.sum((x-np.mean(x))**2))
    return slope, syx, factor

slopes = np.zeros(10000)
syxs = np.zeros(10000)
factors = np.zeros(10000)
for i in range(10000):
    slopes[i] = slope_finder(data_num, n_num, a)[0]
    syxs[i] = slope_finder(data_num, n_num, a)[1]
    factors[i] = slope_finder(data_num, n_num, a)[2]

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(2,1,1)
ax.hist(slopes,50)
ax.set_xlabel('Slope')
ax.set_ylabel('Occurence')
ax.set_title('Histogram of slopes made from random variable y(x)')
ax = fig.add_subplot(2,1,2)
ax.hist(syxs,50)
ax.set_xlabel('S_y|x')
ax.set_ylabel('Occurence')
ax.set_title('Histogram of S_y|x from random variable y(x)')
```

The slope histogram is centered at 1 and has sharply damped tails. On the other hand, the S_y|x histogram has a larger relative full width half max as well as tail regions which appear more "linear".

**2** Assuming $s_{y|x}=1.$ (which is the same as our value of $n$), plot the student-t probability density function on the histogram of b and show that they compare favorably (if they don't you might need to check your work!)

<hr>

```python
import matplotlib.mlab as mlab
import scipy.stats as stats

student_t_range = np.linspace(-4.,4.,len(slopes))
student_t = stats.t.pdf(student_t_range,8)
    
#To compare the two, we subtract by the mean and multiply by the factor given in Lecture 2
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.hist((slopes-np.mean(slopes))*factors,50,normed=True,label = 'slope histogram')
ax.set_xlabel('Slope')
ax.set_ylabel('PDF')
ax.set_title('Histogram of slopes made from random variable y(x) follow student-t distribution')
ax.plot(student_t_range,1.0*student_t,linewidth=2, color='r', label = 'student-t fit, df = 8')
ax.legend(fontsize='small')
```

## Q3: Linear comparison of Deep Cove data to James Bay temperature data

```python
# Load the data:

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]
```

**1** Make a scatterplot of Plot James Bay versus Deep Cove.

```python
import numpy.ma as ma

#the nan elements are transformed to masked elements such that if either the jb
#or the dc data contains a nan, the elements of both from both dc and jb become
#masked elements

m_dc = ma.masked_array(dc, mask=np.isnan(dc+jb))
m_jb = ma.masked_array(jb, mask=np.isnan(dc+jb))

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.scatter(m_dc, m_jb)
ax.set_xlabel('Deep Cove Temperature [Degrees Celsius]')
ax.set_ylabel('James Bay Temperature [Degrees Celsius]')
ax.set_title('Comparison between hourly temperature readouts of Deep Cove vs. James Bay')
```

**2** Compute the linear Correlation Co-efficient.

```python
num_points = len(m_jb) - np.count_nonzero(np.isnan(dc+jb)*1)
sxy = np.sum((m_dc-np.mean(m_dc))*(m_jb-np.mean(m_jb)))/(num_points-1)
rxy = sxy/(np.std(m_jb)*np.std(m_dc))
print rxy
```

**3** Fit a straight line to the James Bay data as a fucntion of the Deep Cove data.  Comment on the fit.  (Do **NOT** use a canned routine for the fitting)

<hr>

```python
b = np.sum(m_dc*m_jb)-num_points*np.mean(m_dc)*np.mean(m_jb)
b = b/(np.sum(m_dc**2)-num_points*np.mean(m_dc)**2)
a = np.mean(m_jb)-b*np.mean(m_dc)
m_jb_hat=b*m_dc+a

fig,ax=plt.subplots(1,1)
ax.plot(m_dc,m_jb,'.')
ax.plot(m_dc,m_jb_hat,'-',label='Fit: (JB)=%1.4f+%1.4f(DC)'%(a,b))
plt.legend(loc=0)
ax.set_xlabel('Deep Cove [C]');ax.set_ylabel('James Bay [C]');
```

The fit displays a strong degree of linearity except for a slight overestimation at 20-25 degrees Celsius and a noticeable deviance in the negative temperatures.

**4** Add the confidence intervals to your fit, and comment.  ***NO CANNED ROUTINES***

```python
# get the error bounds...
syx = np.sqrt(np.sum((m_jb-m_jb_hat)**2)/(num_points-2))
bdn,bup=stats.t.interval(0.95,df=num_points-2,scale=syx)
bb=np.array([bdn,bup])
bb=bb/np.sqrt(np.sum((m_dc-np.mean(m_dc))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])
xx=np.sort(m_dc)
m_jb_hat_high=(b+bb[1])*(xx-np.mean(m_dc))+np.mean(m_jb)
m_jb_hat_low=(b+bb[0])*(xx-np.mean(m_dc))+np.mean(m_jb)
ax.fill_between(xx,m_jb_hat_high,m_jb_hat_low,alpha=0.4)
```

The slope is well determined even with the visual discrepancies seen in the scatter plot. If percentage probability was decreased, we would expect the confidence interval to be even tighter.

**5** Note that we have assumed that the Deep Cove data is independent and that the James Bay data is dependent.  There is no reason to assume this!  Do the fit in the other direction, and compare the slopes.

```python
b = np.sum(m_jb*m_dc)-num_points*np.mean(m_jb)*np.mean(m_dc)
b = b/(np.sum(m_jb**2)-num_points*np.mean(m_jb)**2)
a = np.mean(m_dc)-b*np.mean(m_jb)
m_dc_hat=b*m_jb+a

fig,ax=plt.subplots(1,1)
ax.plot(m_jb,m_dc,'.')
ax.plot(m_jb,m_dc_hat,'-',label='Fit: (DC)=%1.4f+%1.4f(JB)'%(a,b))
plt.legend(loc=0)
ax.set_xlabel('James Bay [C]');ax.set_ylabel('Deep Cove [C]');

# get the error bounds...
syx = np.sqrt(np.sum((m_dc-m_dc_hat)**2)/(num_points-2))
bdn,bup=stats.t.interval(0.95,df=num_points-2,scale=syx)
bb=np.array([bdn,bup])
bb=bb/np.sqrt(np.sum((m_jb-np.mean(m_jb))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])
xx=np.sort(m_jb)
m_dc_hat_high=(b+bb[1])*(xx-np.mean(m_jb))+np.mean(m_dc)
m_dc_hat_low=(b+bb[0])*(xx-np.mean(m_jb))+np.mean(m_dc)

fig,ax=plt.subplots(1,1)
ax.fill_between(xx,m_dc_hat_high,m_dc_hat_low,alpha=0.4)
```

If the relationship was *perfectly* linear, we would expect the slopes to be inverses of each other. However, taking the inverse of the 95% confidence intervals for the James Bay independent data set, we get [0.809,0.815], whose range falls completely out of the 95% confidence interval for the Deep Cove independent data set of [0.754,0.759]. This corroborates with the <1 linear correlation coeffecient.

**NOTE** There is a way to treat both axes independently called "Neutral Regression".  It is relatively simple, and involves normalizing the total distance of each point from the trend line. <a href="http://journals.ametsoc.org/doi/pdf/10.1175/1520-0426(1999)016%3C0876%3AAPFANR%3E2.0.CO%3B2">(Marsden 1999)</a>. This technique is strongly preferred if there is no reason to think that one data set is dependent on the other.  A caveat is that the data sets shoudl be normalized before fitting so that the "distance" has meaning.
