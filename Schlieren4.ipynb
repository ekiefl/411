# Lunesberg Lens

```python
h = 0.1

#calculates angle between two vectors
def angle_bw(v1, v2):
    angle = np.arccos(np.dot(v1, v2))
    if np.isnan(angle):
        if (v1 == v2).all():
            return 0.0
        else:
            return np.pi
    return angle

#returns refractive index at (x,y,z)
def ior(x,y,z):
    if x**2+y**2+z**2>1:
        return 1.5
    else:
        return 1.5*np.sqrt(2-(x**2+y**2+z**2))
    
#returns refractive index gradient at (x,y,z)
def g_ior(x,y,z):
    if x**2+y**2+z**2>1:
        return 0,0,0
    else:
        return -1.5*x/np.sqrt(2-(x**2+y**2+z**2)), -1.5*y/np.sqrt(2-(x**2+y**2+z**2)), -1.5*z/np.sqrt(2-(x**2+y**2+z**2))

#empty lists from which the ensemble of (X,Y) will be filled
x_f = []
y_f = []

'''
for each ray whose initial coordinates are defined by u and v, iterates algorithm (3) in report until z = 1.
'''
inits = np.linspace(-2.5,2.5,140)
for u in range(0,len(inits)):
    for v in range(0,len(inits)):

        #creating ray direction vectors (vxs,vys,vzs) and positional coordinates (xs, ys, and zs)
        vxs = np.zeros(1000)
        vys = np.zeros(1000)
        vzs = np.zeros(1000)
        xs = np.zeros(1000)
        ys = np.zeros(1000)
        zs = np.zeros(1000)

        #initial coordinates and direction
        a = 1
        vxs[0] = 0
        vys[0] = 0
        vzs[0] = 1
        xs[0] = inits[u]*a
        ys[0] = inits[v]*a
        zs[0] = -2*a

        #finds vxs,vys,vzs,xs,ys,zs from previous state
        #breaks once z reaches 1
        for n in range(1,1000):

            if np.floor(zs[n-1]) == 1:
                N = n
                break
            i = ior(xs[n-1],ys[n-1],zs[n-1])
            g_x,g_y,g_z = g_ior(xs[n-1],ys[n-1],zs[n-1])

            vxs[n] = vxs[n-1] + h*g_x/i
            vys[n] = vys[n-1] + h*g_y/i
            vzs[n] = vzs[n-1] + h*g_z/i

            vxs[n] = vxs[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)
            vys[n] = vys[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)
            vzs[n] = vzs[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)

            xs[n] = xs[n-1] + h*vxs[n]
            ys[n] = ys[n-1] + h*vys[n]
            zs[n] = zs[n-1] + h*vzs[n]

        #gets rid of tacked on zeros
        xs = xs[:N]
        ys = ys[:N]
        zs = zs[:N]
        vxs = vxs[:N]
        vys = vys[:N]
        vzs = vzs[:N]
        
        #if s_0 and s_N vary by an angle less than 7.5, the (X,Y) is recorded. Otherwise it is discarded
        if 57.295*np.abs(angle_bw(np.array([vxs[-1],vys[-1],vzs[-1]]), np.array([vxs[0],vys[0],vzs[0]]))) < 7.5:
            x_f.append(xs[-1])
            y_f.append(ys[-1])

#plots Y vs. X, the termination coordinates
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.scatter(x_f,y_f,s=1)
```

```python
'''
X and Y are the grid over which the continuous kernel density estimate is defined. This piece of code takes
x_f and y_f and plots the density. The shape is sharply determined by the parameter bw_method. Complete 
documentation at
http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.gaussian_kde.html
'''
x_f = np.asarray(x_f)
y_f = np.asarray(y_f)
xmin = x_f.min()
xmax = x_f.max()
ymin = y_f.min()
ymax = y_f.max()

X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
values = np.vstack([x_f, y_f])
kernel = stats.gaussian_kde(values, bw_method=0.05)
I = np.reshape(kernel(positions).T, X.shape)


fig = plt.figure(figsize=(7.9,7.9))
ax = fig.add_subplot(111)
test = ax.imshow(np.rot90(I), cmap=plt.cm.copper,
          extent=[xmin, xmax, ymin, ymax])

plt.tight_layout()
ax.set_ylim(-2,2)
ax.set_xlim(-2,2)
plt.xticks(np.arange(-2, 2+1, 1))
plt.yticks(np.arange(-2, 2+1, 1))
ax.set_ylabel('$Y$ [a.u.]')
ax.set_xlabel('$X$ [a.u.]')
```

# Cross Section of Lunesberg Lens

```python

fig = plt.figure(figsize=(12.6,10))
ax = fig.add_subplot(111)

h = 0.005

#returns refractive index at (x,y,z)
def ior(x,y,z):
    if x**2+y**2+z**2>1:
        return 1.5
    else:
        return 1.5*np.sqrt(2-(x**2+y**2+z**2))
    
#returns refractive index gradient at (x,y,z)
def g_ior(x,y,z):
    if x**2+y**2+z**2>1:
        return 0,0,0
    else:
        return -1.5*x/np.sqrt(2-(x**2+y**2+z**2)), -1.5*y/np.sqrt(2-(x**2+y**2+z**2)), -1.5*z/np.sqrt(2-(x**2+y**2+z**2))

#initializes an exiting height list    
exit_error = []


#same as above, except now y_0 is set to 0 so only iterations in x (u variable) occur
inits = np.linspace(-1.5,1.5,20)
for u in range(0,len(inits)):
    #creating ray direction vectors (ss) and positional coordinates (xs, ys, and zs)
    vxs = np.zeros(100000)
    vys = np.zeros(100000)
    vzs = np.zeros(100000)
    xs = np.zeros(100000)
    ys = np.zeros(100000)
    zs = np.zeros(100000)

    a = 1
    #initial coordinates and direction
    vxs[0] = 0
    vys[0] = 0
    vzs[0] = 1
    xs[0] = inits[u]*a
    ys[0] = 0
    zs[0] = -1.5*a

    #finds vxs,vys,vzs,xs,ys,zs from previous state
    #breaks once z reaches 1.5
    for n in range(1,10000):

        if zs[n-1] > 1.5:
            N = n
            break
        i = ior(xs[n-1],ys[n-1],zs[n-1])
        g_x,g_y,g_z = g_ior(xs[n-1],ys[n-1],zs[n-1])

        vxs[n] = vxs[n-1] + h*g_x/i
        vys[n] = vys[n-1] + h*g_y/i
        vzs[n] = vzs[n-1] + h*g_z/i

        vxs[n] = vxs[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)
        vys[n] = vys[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)
        vzs[n] = vzs[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)

        xs[n] = xs[n-1] + h*vxs[n]
        ys[n] = ys[n-1] + h*vys[n]
        zs[n] = zs[n-1] + h*vzs[n]

    #appends the exiting height to the exit_error list    
    exit_error.append(xs[N-1])
    
    #gets rid of tacked on zeros
    xs = xs[:N]
    ys = ys[:N]
    zs = zs[:N]
    vxs = vxs[:N]
    vys = vys[:N]
    vzs = vzs[:N]
    
    ax.plot(zs,xs,color='k')
    ax.set_xlim(-1.5,1.5)
    ax.set_ylim(-1.5,1.5)

    
#omits exit_heights if ray didn't go through lens
exit_error = np.asarray(exit_error)
exit_error=exit_error[exit_error<1.]
exit_error=exit_error[exit_error>-1.]
print exit_error
print 'exit heights are ', np.mean(np.abs(exit_error))

    
'''
plots the results and creates a color map of the refractive index
'''
circle1=plt.Circle((0,0),1,color='k',Fill=False,linestyle='dashed',alpha=0.4)
ax.add_artist(circle1)
xx, zz = np.mgrid[-1.5:1.5:400j,-1.5:1.5:400j]
nn = 1.5*np.sqrt(2-(xx**2+zz**2))

#defines refractive index to be 1.5 if outside the lens
for i in range(0,400):
    for j in range(0,400):
        if xx[i][j]**2+zz[i][j]**2>1:
            nn[i][j] = 1.5

cmap = plt.get_cmap('Blues') 
pcm=ax.pcolormesh(xx,zz,nn, cmap=cmap)
cbar = plt.colorbar(pcm)
cbar.set_label('Refractive Index')
ax.set_xlabel('$z$   [a.u.]')
ax.set_ylabel('$y$   [a.u.]')
```

# Gaussian Cylinder

```python
'''
exact same procedure as for the luneburg lens, except ior and g_ior are redefined
'''

def ior(x,y,z):
    return 1.00027 - 0.00017*np.exp(-(x**2+z**2))
def g_ior(x,y,z):
    return -0.0001*2*x*np.exp(-(x**2+z**2)), 0., -0.0001*2*z*np.exp(-(x**2+z**2))

h = 0.1
x_f = []
y_f = []
ang = []
r = []

inits = np.linspace(-4.5,4.5,80)
for u in range(0,len(inits)):
    for v in range(0,len(inits)):
        vxs = np.zeros(100)
        vys = np.zeros(100)
        vzs = np.zeros(100)
        xs = np.zeros(100)
        ys = np.zeros(100)
        zs = np.zeros(100)

        a = 1
        vxs[0] = 0
        vys[0] = 0
        vzs[0] = 1
        xs[0] = inits[u]*a
        ys[0] = inits[v]*a
        zs[0] = -2*a

        for n in range(1,1000):
            #depdends on length
            if n == 999:
                N = n
                break
            if np.floor(zs[n-1]) == 3:
                N = n
                break
            i = ior(xs[n-1],ys[n-1],zs[n-1])
            g_x,g_y,g_z = g_ior(xs[n-1],ys[n-1],zs[n-1])

            vxs[n] = vxs[n-1] + h*g_x/i
            vys[n] = vys[n-1] + h*g_y/i
            vzs[n] = vzs[n-1] + h*g_z/i

            vxs[n] = vxs[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)
            vys[n] = vys[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)
            vzs[n] = vzs[n]/np.sqrt(vxs[n]**2 + vys[n]**2 + vzs[n]**2)

            xs[n] = xs[n-1] + h*vxs[n]
            ys[n] = ys[n-1] + h*vys[n]
            zs[n] = zs[n-1] + h*vzs[n]


        xs = xs[:N]
        ys = ys[:N]
        zs = zs[:N]
        vxs = vxs[:N]
        vys = vys[:N]
        vzs = vzs[:N]
        
        eps = 57.295*np.abs(angle_bw(np.array([vxs[-1],vys[-1],vzs[-1]]), np.array([vxs[0],vys[0],vzs[0]])))
        ang.append(eps)
        r.append(np.abs(inits[u]))
        
        #velocity discriminating
        if eps < 0.004:
            x_f.append(xs[-1])
            y_f.append(ys[-1])
        
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.scatter(x_f,y_f,s=1)
```
