## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline

N = 2048
x = np.random.randn(N)
num_bin = 8

#making bins for histograms
a = 50
m_bin = np.linspace(0.,10.,a+1)

Graw,fraw = mlab.psd(x,NFFT=N,Fs=1,window=mlab.window_none,noverlap=0)
Graw = Graw/2. #normalize by E(Gxx)
Gbox,fbox = mlab.psd(x,NFFT=N/num_bin,Fs=1,window=mlab.window_none,noverlap=0)
Gbox = Gbox/2.
Ghann,fhann = mlab.psd(x,NFFT=N/num_bin,Fs=1,window=mlab.window_hanning,noverlap=N/num_bin/2)
Ghann = Ghann/2.

r1 = np.linspace(0,20,100)
r2 = np.linspace(0,60,100)
r3 = np.linspace(0,120,100)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

ax.hist(Graw, m_bin, color = 'b',normed=True, histtype='step',linewidth='2',label='single')
ax.hist(Gbox, m_bin, color = 'c', normed=True, histtype='step',label = 'block',linewidth='2')
ax.hist(Ghann, m_bin, normed=True, color='r', histtype='step',label = 'hann',linewidth='2')

ax.plot(r1/2.,stats.chi2.pdf(r1, 2)/np.trapz(stats.chi2.pdf(r2, 2),r2/2), 'b', linewidth=2)
ax.plot(r2/2./num_bin,stats.chi2.pdf(r2, 2*num_bin)/np.trapz(stats.chi2.pdf(r2, 2*num_bin),r2/2./num_bin), 'c',linewidth=2)
ax.plot(r3/(2*num_bin-1)/2.,stats.chi2.pdf(r3, 
                                           2*(2*num_bin-1))/np.trapz(stats.chi2.pdf(r3, 2*(2*num_bin-1)),r3/2./(2*num_bin-1)),
        'r',linewidth=2)

ax.set_xlabel('Power Spectrum [V^2/Hz]')
ax.set_ylabel('Normalized Occurence')
ax.set_title(r'The Distributions for $N_{B}$-blocked Power Spectra follow $\chi^{2}_{2N_{B}}$ Distributions')
ax.set_xlim(0,6)
ax.legend()

'----------------------------------------'

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.set_xlim(0,5)
ax.set_xlabel('Power Spectrum [V^2/Hz]')
ax.set_ylabel('Normalized Occurence')
ax.set_title('The distribution tightens as block number increases for Box-Car Distribution')

Ghann8,fhann = mlab.psd(x,NFFT=N/8,Fs=1,window=mlab.window_hanning,noverlap=N/8/2)
Ghann32,fhann = mlab.psd(x,NFFT=N/32,Fs=1,window=mlab.window_hanning,noverlap=N/32/2)

var8 = np.var(Ghann8)
var32 = np.var(Ghann32)

ax.hist(Ghann8, 25, normed=True, histtype='step',linewidth=2, label='Bin = 8; Var = %f' %(var8))
ax.hist(Ghann32, 25, normed=True, histtype='step',linewidth=2, label='Bin = 32; Var = %f' %(var32))
ax.legend()
```

We see that the single segment calculation does indeed follow a chi2 distribution with df = 2 which is broadly spread and has a most probable value of 0. On the otherhand, hanning methods as well as boxcar methods each lead to chi2 distributions with df = 2n, where n is the number of blocks. These show to have much smaller variances, and are thus more accurate estimates of the power spectra, the hanning method being an even tighter distribution due to spectral leakage minimization and increased degrees of freedom.

This relationship between number of blocks and the variance of the distribution is depicted in the 2nd figure, which shows that increasing the number of blocks decreases the variance.

<div style='background:#F0F0F0'>**2**  For the Hanning window estimate, what fits better $2N_{blocks}$ degrees of freedom or $18N_{blocks}/11$ degrees of freedom (the correct theoretical value)?  Its OK to evaluate by eye.</div>

```python
Ghann,fhann = mlab.psd(x,NFFT=N/num_bin,Fs=1,window=mlab.window_hanning,noverlap=N/num_bin/2)
Ghann = Ghann/2. #normalize by E(Gxx)

r3 = np.linspace(0,120,300)
pdf_2N = stats.chi2.pdf(r3, 2*num_bin)
pdf_2N = pdf_2N/np.trapz(stats.chi2.pdf(r3, 2*num_bin),r3/2./num_bin)
pdf_1811N = stats.chi2.pdf(r3, 18./11*num_bin)
pdf_1811N = pdf_1811N/np.trapz(stats.chi2.pdf(r3,18./11*num_bin),r3/2./num_bin)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.set_xlim(0,5)
ax.set_xlabel('Power Spectrum [V^2/Hz]')
ax.set_ylabel('Normalized Occurence')
ax.set_title('Comparison between df = 2 and df = 18/11')

ax.hist(Ghann, m_bin, normed=True, color='b', histtype='step',label = 'hann',linewidth='2')
ax.plot(r3/2./num_bin, pdf_2N, label = 'df = 2',linewidth=2)
ax.plot(r3/2./num_bin, pdf_1811N, label = 'df = 18/11',linewidth=2)
ax.legend()

```

Surprisingly, the data is better represented by the df = 2 distribution than the df = 18/11 distribution! This difference indicates our approximation is still not optimized.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]
```

```python
size = 1024

Gbox,fbox = mlab.psd(dc,NFFT=size,Fs=1./3600,window=mlab.window_none,noverlap=0)
Ghann,fhann = mlab.psd(dc,NFFT=size,Fs=1./3600,window=mlab.window_hanning,noverlap=size/2)

fig = plt.figure(figsize=(8,15))
ax = fig.add_subplot(2,1,1)
ax.loglog(fbox,Gbox,label='box',linewidth=2)
ax.loglog(fhann,Ghann,label='hanning',linewidth=2)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Spectrum [C^2/Hz]')
ax.set_title('Comparison between Hanning window and Boxcar Method for Deep Cove Data')
ax.set_xlim(0,np.max(fbox))
ax.set_ylim(1e2,1e9)
ax.legend()
```

Both the boxcar and the hanning method are capable of capturing the often subtle peaks throughout the data. However, there is an order of magnitude difference between the boxcar and hanning methods for the first data point, which is probably due to spectral leakage. The boxcar method is incapable of measuring the peak occuring near the first data point.

<div style='background:#F0F0F0'> **2** Now do the same thing to the intergal of the Deep Cove data with time: $$y(t)=\int_o^t x(t')\ \mathrm{d}t'$$  This is obviously a silly thing to do, but compare the spectra and comment on the difference between the Hanning window and the non-Hanning windowed data.  </div>

```python
dc_int = np.zeros(len(dc))
for el in range(0,len(dc)):
    dc_int[el] = np.trapz(dc[:el+1])

Gbox,fbox = mlab.psd(dc_int,NFFT=size,Fs=1./3600.,window=mlab.window_none,noverlap=0)
Ghann,fhann = mlab.psd(dc_int,NFFT=size,Fs=1./3600,window=mlab.window_hanning,noverlap=size/2)

time = np.arange(0,3600*len(dc),3600)
fig = plt.figure(figsize=(8,15))
ax = fig.add_subplot(2,1,1)
ax.plot(time,dc_int)
ax.set_xlabel('Time [s]')
ax.set_ylabel('[C s]')
ax.set_title('Integrated Temperature Data for Deep Cove')

ax = fig.add_subplot(2,1,2)
ax.loglog(fbox,Gbox,label='box',linewidth=2)
ax.loglog(fhann,Ghann,label='hanning',linewidth=2)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Spectrum [(C s)^2/Hz]')
ax.set_title('Power Spectra of integrated Deep Cove temperature data')
ax.set_xlim(0,np.max(fbox))
ax.legend()
```

For the integrated Deep Cover weather data, the boxcar and hanning methods bear no resemblance. Even after the integration, the hanning method is still capable of fully resolving the peaks that were observed in the original data, whereas the boxcar method has failed miserably. From this we conclude that for such a signal, the hanning method is much more suited.
