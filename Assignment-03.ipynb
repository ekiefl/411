## Q1: Lag correlations of weather time series

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc_raw = hourdata[0,:]
jb_raw =hourdata[1,:]
````

**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xy}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememebr to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
m_dc = ma.masked_array(dc, mask=np.isnan(dc))
m_jb = ma.masked_array(jb, mask=np.isnan(jb))

rho_xx = np.zeros(24*30+1)
tau_list = np.arange(0,24*30+1)

rho_xx[0] = np.var(m_dc)/np.var(m_dc)
for tau in range(1,len(tau_list)):
    rho_xx[tau] = np.mean((m_dc[:-tau]-np.mean(m_dc))*(m_dc[tau:]-np.mean(m_dc[tau:])))/np.var(m_dc)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.plot(tau_list,rho_xx)
ax.set_xlabel('Lag time [h]')
ax.set_ylabel(r'$\rho_{xy}$',fontsize=20)
ax.set_title('Lag correlation co-efficient for Deep Cove temperature')
ax.set_ylim((0.43,1.))
````

The self-correlation has slow decay which is occuring over a time scale on the order of 1000 hours, which might be due to a gradually changing weather system. This decay is also being modulated by periodic daily fluctuations. Since the data starts at midnight, the periodic fluctuation peaks correspond to night time temperatures because they are more correlated, whereas the troughs correspond to day-time temperatures.

*2* Now, instead of doing the whole year, look at just the winter months and the summer months and compare the lag correlations (no need to do this for the whole time series, just choose one winter and one summer).  Make sure to indicate which lag correlation is which on the plots.

Comment below on the different charcter of these two lag correlations.

````python
# If its not a leap year, the start and end of summer are the 172nd and 265th 
# days of the Julian Calendar. Likewise, the start and end of winter are the
# 355th and 79th days of the Julian Calendar. This justifies the indexing chosen

# the masked array is useful between it retains the indexing

S = m_dc[24*172:24*265]
W = m_dc[24*355:24*(365+79)]

rho_xx_S = np.zeros(len(S))
tau_list_S = np.arange(0,len(S))

rho_xx_W = np.zeros(len(W))
tau_list_W = np.arange(0,len(W))

rho_xx_S[0] = np.var(S)/np.var(S)
rho_xx_W[0] = np.var(W)/np.var(W)

for tau in range(1,len(tau_list_S)):
    rho_xx_S[tau] = np.mean((S[:-tau]-np.mean(S))*(S[tau:]-np.mean(S[tau:])))/np.var(S)
    
for tau in range(1,len(tau_list_W)):
    rho_xx_W[tau] = np.mean((W[:-tau]-np.mean(W))*(W[tau:]-np.mean(W[tau:])))/np.var(W)
    

fig = plt.figure(figsize=(8,12))
ax = fig.add_subplot(2,1,1)
ax.plot(tau_list_S,rho_xx_S,'r')
ax.set_xlabel('Lag time after Summer Solstice [h]')
ax.set_ylabel(r'$\rho_{xy}$',fontsize=20)
ax.set_title('Lag correlation co-efficient for Deep Cove temperature during Summer')
ax = fig.add_subplot(2,1,2)
ax.plot(tau_list_W,rho_xx_W)
ax.set_xlabel('Lag time after Winter Solistice [h]')
ax.set_ylabel(r'$\rho_{xy}$',fontsize=20)
ax.set_title('Lag correlation co-efficient for Deep Cove temperature during Winter')

````

The Summer lag correlation has constant long term behaviour with large daily fluctuations arising from night and day, whereas the Winter lag correlation has dynamic long term behaviour and relatively small daily fluctuations arising from night and day. I interpret this to mean that a) Summer weather systems are less dynamically changing than in Winter and b) temperature differences between night and day are more significant in Summer than they are in Winter.

### Proving that the numpy.fft does it's job

To demonstrate that numpy.fft works, I am going to bury some sinusoidal signals in the midst of random noise and verify fft peaks at said frequencies.

````python
N = 2000

t = np.linspace(0,1,N)
x = 30*np.random.randn(N)
s1 = 10.*np.cos(800*2*np.pi*t)
s2 = 10*np.cos(300*2*np.pi*t)
s3 = 20*np.sin(200*2*np.pi*t)
x = x + s1 + s2 + s3

psd = np.fft.fft(x)*np.conjugate(np.fft.fft(x))
psd = np.real(psd)
psd = psd[:len(psd)/2.]
f = np.linspace(1,N/2.,N/2.)

fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(2,1,1)
ax.plot(f,psd)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Spectrum')
ax.set_title('Sinusoidal frequencies are extractable from the noise')
ax = fig.add_subplot(2,1,2)
ax.plot(t,x)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.set_title('Subsample of Original Signal')
ax.set_xlim(0,0.1)


````

````python
### Method 1 #### Calculating PSD with |Fourier(x)|^2 instead of Fourier(Rxx)

def PSD_calc(T):
    
    b = 100 #as before
    y = np.random.randn(T+b)
    x = np.zeros(T)
    #constructing x from y
    for i in range(0,T):
        x[i] = np.trapz(y[i:i+b+1])
        
    Sxx = np.fft.fft(x)*np.conjugate(np.fft.fft(x))
    Sxx = np.real(Sxx) #just to get rid of all the +0. j's 
    Sxx = Sxx[:len(Sxx)/2.]
    return Sxx

b = 100
freq_theor = np.linspace(1e-3,0.5,100000)
PSD_theor = (-1./2)*(np.cos(2*np.pi*b*freq_theor)-1)/(np.pi**2*freq_theor**2)

PSD_5000 = PSD_calc(5000)
freq_5000 = np.linspace(1e-3,0.5,2500)

PSD_50000 = PSD_calc(50000)
freq_50000 = np.linspace(1e-3,0.5,25000)

PSD_500000 = PSD_calc(500000)
freq_500000 = np.linspace(1e-3,0.5,250000)

fig = plt.figure(figsize=(12,36))
ax = fig.add_subplot(4,1,1)
ax.loglog(freq_theor,PSD_theor,'--',label='Theory')
ax.loglog(freq_500000,PSD_500000,'c', label='T=500000')
ax.loglog(freq_50000,PSD_50000, label='T=50000')
ax.loglog(freq_5000,PSD_5000,label='T=5000')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Spectrum [V^2]')
ax.set_title('The data shares the same qualitative behaviour as the theory')
ax.set_xlim(1e-3,0.1)
ax.legend(loc='best')
````

The results are offset by the theoretical curve by about 4-5 orders of magnitude, which I suspect is due to my own error. Regardless, it *is* seen that at higher T values, the spectrum artificially rises and the noise level grows which shows that it is not converging to anything.
