```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline

N = 2048
x = np.random.randn(N)
num_bin = 8

#making bins for histograms
a = 50
m_bin = np.linspace(0.,10.,a+1)

#boxcar psd calc
Gbox,fbox = mlab.psd(x,NFFT=N/num_bin,Fs=1,window=mlab.window_none,noverlap=0)

#plotting
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.hist(Gbox, m_bin, color = 'c', normed=True, histtype='step',label = 'block',linewidth='2')


#if we scale the x axis of the pdf, we must also renormalize so that the area is 1. This is what happens here
r = np.linspace(0,60,100)
unscaled = stats.chi2.pdf(r, 2*num_bin)
scaled_by_n = unscaled/np.trapz(stats.chi2.pdf(r, 2*num_bin),r/num_bin)
scaled_by_nu = unscaled/np.trapz(stats.chi2.pdf(r, 2*num_bin),r/num_bin/2.)

#now we scale by x and plot
ax.plot(r, unscaled, label = 'unscaled',linewidth=4)
ax.plot(r/num_bin, scaled_by_n, label = 'scaled by n',linewidth=4)
ax.plot(r/num_bin/2, scaled_by_nu, label = 'scaled by nu',linewidth=4)
ax.set_xlim(0,10)
ax.legend()
```
