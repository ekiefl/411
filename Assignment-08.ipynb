### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
k
N=1e3
dt = 0.1
x0 = np.random.randn(N)
t0 = np.arange(0,N*dt,dt)

#Bandwidth = 0.5 Hz
B=0.5
# filter to make Bandlimited signal
fp=B*0.7
fs = B*0.9
fNyq=1/2./dt
n,fn=signal.ellipord(fp/fNyq, fs/fNyq,0.1, 60.)
b,a=signal.ellip(n,0.1,60.,fn)
xBL = signal.filtfilt(b,a,x0)

nfft = 10240
GxBL,fxBL = mlab.psd(xBL,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2.)
Gx0,fx0 = mlab.psd(x0,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2.)


fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(1,1,1)
ax.loglog(fxBL,GxBL,label='Band-limited')
ax.loglog(fx0,Gx0,label='Unfiltered')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Gxx [V^2 Hz^-1]')
ax.legend(loc = 'best')
ax.set_ylim(1e-10,1)
```

<div style='background:#F0F0F0'>**2** Now decimate by a factor of 4 to get a time series sampled at 2.5 Hz.  Show that you can recover the 10 Hz time series using the Sampling Theorem.</div>

```python
#2.5

tn = t0[::4]
xn = xBL[::4]
Ns = 250
fig,ax=plt.subplots(1,1)
n = np.arange(Ns)
xreco=1.*xBL[:Ns*4]
for i in range(Ns*4):
    xreco[i] = np.sum(xn[:Ns]*np.sinc(2*B*t0[i]-n))
ax.plot(t0[:Ns*4],(xreco-xBL[:Ns*4]),label='reconstruction')
ax.plot(t0,-xBL,label='original')
plt.xlabel('t [s]')
plt.ylabel('$(x_{recon}-x(t))$')
plt.legend(fontsize='small')
ax.set_xlim(0,100)
```

The signals qualitative is completely retained however many of the high peaks are underestimated and many of the low peaks are overestimated. It also does appear to be missing some low frequency information since at ~75s the difference bewteen the two is almost nothing, at ~87.5 the difference is maximal, and by ~100s the difference is, again, almost nothing. This is counterintuitive that low frequency information is not retained by applying the sampling theorem. Could this be the effect of aliasing?

<div style='background:#F0F0F0'>**3** Now decimate by a factor of 40 to get a time series sampled at 0.25 Hz.  Show that you cannot recover the 10 Hz time series using the Sampling Theorem.</div>

```python
tn = t0[::40]
xn = xBL[::40]
Ns = 25
fig,ax=plt.subplots(1,1)
n = np.arange(Ns)
xreco=1.*xBL[:Ns*40]
for i in range(Ns*40):
    xreco[i] = np.sum(xn[:Ns]*np.sinc(2*B*t0[i]-n))
ax.plot(t0[:Ns*40],(xreco-xBL[:Ns*40]),label='reconstruction')
ax.plot(t0,-xBL,label='original')
plt.xlabel('t [s]')
plt.ylabel('$(x_{recon}-x(t))$')
plt.legend(fontsize='small')
ax.set_xlim(0,40)
```

The reconstruction is clearly less convergent from t less than 20 than in the last case, however for t greater than 20 there is suddenly perfect convergence. This is an artifact that arises due to the nature of sinc near 0: sinc(0)=1 so clearly the argument is becoming *too* close to 1 which explains this artificial convergence.

## Q2: Cubic Spline

Using the data below...

```python
from scipy.interpolate import interp1d

t = np.array([0.,2.5,2.7,3.4,4.2,6.5,9.0])
x = np.cos(t*2*np.pi/3)
```

<div style='background:#F0F0F0'>**1** Perform a cubic spline interpolation onto an even grid from 0 to 9.0 by 
  1. creating the matrix from the notes
  2. inverting and solving for the cubic co-efficients
  3. using those polynomials to evaluate the interpolated data.
Compare with using `scipy.interpolate.interp1d`.
Comment on any similarities or differences between the two results.</div>

```python
#defining all the Deltas in terms of arrays
d = np.zeros(6)
for i in range(0,len(d)):
    d[i] = t[i+1]-t[i]
dx = np.zeros(6)
for i in range(0,len(d)):
    dx[i] = x[i+1]-x[i]
 
#creating the matrix
M = np.zeros((len(t),len(t)))
M[0,0] = 1; M[-1,-1] =1
M[1,0] = d[0]
M[5,6] = d[5]
for i in range(1,5):
    M[i+1,i] = d[i]; M[i,i+1] = d[i]
for i in range (1,6):
    M[i,i] = 2*(d[i-1]+d[i])
    
#matrix product is xx
xx = np.zeros(7)
for i in range(1,6):
    xx[i] = dx[i]/d[i]-dx[i-1]/d[i-1]
xx = 3*xx

#calculating the inverse to solve for c
C = np.dot(np.linalg.inv(M),xx)

#solving other coeffecients with C
D = np.zeros(7)
B = np.zeros(7)
A = x
for i in range(0,5):
    D[i] = (C[i+1]-C[i])/3./d[i]
for i in range(0,6):
    B[i] = (dx[i]-d[i]**2*C[i]-d[i]**3*D[i])/d[i]

#linspace for which the function q(t) will be defined
T = np.linspace(0,9.0,1000)
bdy = np.zeros(len(t))

#finds the indices of T values that most closely resemble the t values
for j in range(0,len(t)):
    time = t[j]
    for i in range(0,len(T)-2):    
        if T[i]<=time and T[i+1]>time:
            bdy[j] = i
        bdy[-1]=len(T)-1

#I'm sorry that the following code is ugly, but its conceptually simple: I'm plotting A + BT + CT^2 + DT^2
#for intervals between the values in t, where A,B,C,D change index depending on which interval is being considered.
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(1,1,1)
ax.plot(t,x,label='actual data')
ax.plot(T[bdy[0]:bdy[1]],
        A[0]+B[0]*(T[bdy[0]:bdy[1]]-T[bdy[0]])+C[0]*(T[bdy[0]:bdy[1]]-T[bdy[0]])**2+D[0]*(T[bdy[0]:bdy[1]]-T[bdy[0]])**3,
        'k--')
ax.plot(T[bdy[1]:bdy[2]], 
        A[1]+B[1]*(T[bdy[1]:bdy[2]]-T[bdy[1]])+C[1]*(T[bdy[1]:bdy[2]]-T[bdy[1]])**2+D[1]*(T[bdy[1]:bdy[2]]-T[bdy[1]])**3,
        'k--')
ax.plot(T[bdy[2]:bdy[3]], 
        A[2]+B[2]*(T[bdy[2]:bdy[3]]-T[bdy[2]])+C[2]*(T[bdy[2]:bdy[3]]-T[bdy[2]])**2+D[2]*(T[bdy[2]:bdy[3]]-T[bdy[2]])**3,
        'k--')
ax.plot(T[bdy[3]:bdy[4]], 
        A[3]+B[3]*(T[bdy[3]:bdy[4]]-T[bdy[3]])+C[3]*(T[bdy[3]:bdy[4]]-T[bdy[3]])**2+D[3]*(T[bdy[3]:bdy[4]]-T[bdy[3]])**3,
        'k--')
ax.plot(T[bdy[4]:bdy[5]], 
        A[4]+B[4]*(T[bdy[4]:bdy[5]]-T[bdy[4]])+C[4]*(T[bdy[4]:bdy[5]]-T[bdy[4]])**2+D[4]*(T[bdy[4]:bdy[5]]-T[bdy[4]])**3,
        'k--')
ax.plot(T[bdy[5]:bdy[6]], 
        A[5]+B[5]*(T[bdy[5]:bdy[6]]-T[bdy[5]])+C[5]*(T[bdy[5]:bdy[6]]-T[bdy[5]])**2+D[5]*(T[bdy[5]:bdy[6]]-T[bdy[5]])**3,
        'k--')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')

#now we use the software package
ff = interp1d(t,x,kind='cubic')
yi=ff(T)
ax.plot(T,yi,label='interp1d cubic spline')
ax.legend(loc = 'best')
```

It's clear that our method (dashed black) for interpolating this data is much more effective than the built in routine in interp1d. This is seen by a much tighter "hug" to the original data, whereas the interp1d cubic spline appears to be grossly overestimating some of the constants. The two methods are similar in the fact that they both tend to share inflection points; this is no doubt related to the demand that they intersect each data point.
