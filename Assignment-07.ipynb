### **Q1** Frequency-domain filtering

<div style='background:#F0F0F0'>  Consider a time series with a sharp jump in the middle as in the notes:
</div>

```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline

N=1001
fs=1.
t=np.arange(0,N/fs,1/fs)

x = np.cumsum(np.random.randn(N)) # red noise
f0=0.03 # Hz
x = x+ 40.*np.sin(f0*2*np.pi*t) # + sine wave
x=x-np.mean(x)
noise = 60.*np.random.randn(N)
xn = x+noise
xn[N/2:]=xn[N/2:]-2000.
# remove the mean!!
xn = xn-np.mean(xn)
```

<div style='background:#F0F0F0'>**1** Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of $f_c=0.06\ Hz$.</div>
   
  1. plot the signal and the smoothed signal (suggest zooming in on the step).
  2. plot the spectral response $|H(f)|$
  3. plot the phase response $\mathrm{angle}(H(f))$.

<div style='background:#F0F0F0'>
Be sure to comment on how good your filter is.
</div>

```python
X = np.fft.fft(xn)
ff = np.linspace(0,fs,N)
fc = 0.06

X[(ff>fc) & (ff<(fs-fc))]=1.e-15
y = np.real(np.fft.ifft(X))+np.mean(xn)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(t,xn,label='Signal')
ax.plot(t,y,linewidth=3,label='Filtered')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.legend(loc = 'best')
ax.set_xlim(400,600)

Gxx,fxx = mlab.psd(xn,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
Gyy,fyy = mlab.psd(y,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
Gxy, fxy = mlab.csd(xn,y,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
H = np.sqrt(Gyy/Gxx)
phi = np.angle(Gxy)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(fxx,H,label='abs(H)')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Gxx [V^2 Hz^-1]')
ax.legend(loc = 'best')

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(fxx,phi,label='Angle(H)')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Angle [Radians]')
ax.legend(loc = 'best')
```

The filter does a generally good job when far away from the step at 500s, However breaks down near the step function. This can be partially explained by the fact that fourier coeffecients of a step function are equally weighted across all frequencies, and since we set the higher frequencies to zero, the filter is unable to reproduce the sharpness of the step function.

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

```python
X = np.fft.fft(xn)
ff = np.linspace(0,fs,N)

q = 0.06/0.001 #index over which the hanning window is to be applied, i.e. (0,0.06)Hz
H_right = 0.5*(1-np.cos(2*np.pi*np.arange(0,q)/(2*q-1)))
H_left = 0.5*(1-np.cos(2*np.pi*np.arange(q,2*q)/(2*q-1)))
Hfull = np.zeros(N)
Hfull[:q] = H_left[:]; Hfull[N-q:] = H_right[:]; Hfull[q:N-q] = 1e-15 #defining the full scaling function
X = Hfull*X
y = np.real(np.fft.ifft(X))+np.mean(xn)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(ff,Hfull,label='Hanning Window')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Scaling Factor')
ax.legend(loc = 'best')

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(t,xn,label='Signal')
ax.plot(t,y,'g',linewidth=3,label='Filtered')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.legend(loc = 'best')
ax.set_xlim(400,600)

Gxx,fxx = mlab.psd(xn,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
Gyy,fyy = mlab.psd(y,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
Gxy, fxy = mlab.csd(xn,y,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
H = np.sqrt(Gyy/Gxx)
phi = np.angle(Gxy)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(fxx,H,label='abs(H)')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Gxx [V^2 Hz^-1]')
ax.legend(loc = 'best')
ax.set_xlim(0,0.2)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(2,1,1)
ax.plot(fxx,phi,label='Angle(H)')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Angle [Radians]')
ax.legend(loc = 'best')
ax.set_xlim(0,0.2)
```

Clearly, the rolloff is much slower than in the first case. Also, the amplitude of the embedded frequency isn't well preserved. This is because the response function at 0.03Hz is ~0.5.

<div style='background:#F0F0F0'>**3** So the Hanning window falls off a *lot* earlier and more slowly than a step function, so adjust the cutoff frequency to get a reasonable filter that preserves the amplitude of the sine wave at $f=0.03\ \mathrm{Hz}$.  Comment on the pros and cons of this approach.  
</div>

```python
X = np.fft.fft(xn)
ff = np.linspace(0,fs,N)

q = 0.13/0.001 #index over which the hanning window is to be applied, i.e. (0,0.06)Hz
H_right = 0.5*(1-np.cos(2*np.pi*np.arange(0,q)/(2*q-1)))
H_left = 0.5*(1-np.cos(2*np.pi*np.arange(q,2*q)/(2*q-1)))
Hfull = np.zeros(N)
Hfull[:q] = H_left[:]; Hfull[N-q:] = H_right[:]; Hfull[q:N-q] = 1e-15 #defining the full scaling function
X = Hfull*X
y = np.real(np.fft.ifft(X))+np.mean(xn)

fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(2,1,1)
ax.plot(ff,Hfull,label='Hanning Window')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Scaling Factor')
ax.legend(loc = 'best')

fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(2,1,1)
ax.plot(t,xn,label='Signal')
ax.plot(t,y,'g',linewidth=3,label='Filtered')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.legend(loc = 'best')
ax.set_xlim(400,600)

Gxx,fxx = mlab.psd(xn,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
Gyy,fyy = mlab.psd(y,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
Gxy, fxy = mlab.csd(xn,y,NFFT=256,Fs=fs,window=mlab.window_hanning,noverlap=128)
H = np.sqrt(Gyy/Gxx)
phi = np.angle(Gxy)

fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(2,1,1)
ax.plot(fxx,H,label='abs(H)')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Gxx [V^2 Hz^-1]')
ax.legend(loc = 'best')

fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(2,1,1)
ax.plot(fxx,phi,label='Angle(H)')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Angle [Radians]')
ax.legend(loc = 'best')
```

Changing the cutoff to 0.13Hz, most of the 0.3Hz signal is preserved (~85%), and simultaneously, the dropoff at 500s is steeper and more respresentative. The tradeoff is that more of the noise signal is left unfiltered.

Recap:

Pros - 0.3Hz amplitude preserved, dropoff at 500s is steeper

Cons - more of the noise signal is left unfiltered

```python
dt = 60.
fs = 1/dt
nyq = fs/2.

#daily signal is f = 1.15740740741e-05 Hz

fp_l = 1e-7
fs_l = 1e-6

#creating a low pass filter and applying it to dc
n_l, fn_l = signal.ellipord(fp_l/nyq, fs_l/nyq, 1., 40.)
b_l, a_l = signal.ellip(n_l, 1., 40., fn_l, btype='lowpass')
y_lo = signal.filtfilt(b_l,a_l,dc)

fp_h = 1e-4
fs_h = 1e-3

#creating a high pass filter and applying it to the low pass filter to create a bandpass
n_h, fn_h = signal.ellipord(fp_h/nyq, fs_h/nyq, 1., 40.)
b_h, a_h = signal.ellip(n_h, 1., 40., fn_h, btype='highpass')
y = signal.filtfilt(b_h,a_h,y_lo)

Gyy,fyy = mlab.psd(y,NFFT=2000,Fs=fs,window=mlab.window_hanning,noverlap=1000)
Gxx,fxx = mlab.psd(dc,NFFT=2000,Fs=fs,window=mlab.window_hanning,noverlap=1000)

fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(2,1,1)
ax.loglog(fyy, Gxx)
ax.loglog(fyy, Gyy)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Gxx [V^2 Hz^-1]')
ax.legend(loc = 'best')
#ax.set_xlim(1*7200,2*7200)
ax.legend()
```
