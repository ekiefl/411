**1** Using the data from the hourly time series at Deep Cove (`dc`), show that the canned `fft` routine returns the correct discrete Fourier Transform of a suitable subset of the data (your choice on what subset).  You can do this graphically, by comparing the amplitude and the phase of the Fourier components, or simply show that the resulting numbers are the same.  

Hints: 
  - ignore bad data (rather than dealing with it "properly").  Make sure the units of your FT are: $\mathrm{^oC\, s}$.
  - If you want to compute the DFT effeciently, form the matrix $F$ referred to in the notes, and plot some of the time series associated with the different entries and make sure they look like Sines and Cosines and that the edge conditions are what you want


````python
dc_sub = dc_raw[100:200]

N = len(dc_sub)
T = N*3600
k = np.arange(0,N/2+1)
X = np.empty(N/2+1,dtype=complex)

for j in range(0,N/2+1):
    real = np.zeros(N)
    imag = np.zeros(N)
    for i in range(0,len(dc_sub)):
        real[i] = T/N*dc_sub[i]*np.cos(2*np.pi*i*j/N)
        imag[i] = T/N*dc_sub[i]*np.sin(2*np.pi*i*j/N)
    X.real[j] = np.sum(real)
    X.imag[j] = -np.sum(imag)

q = 3600*np.fft.fft(dc_sub)
q = q[:len(q)/2+1] # data is symmetric about N/2, except @ f = 0.

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(2,1,1)
ax.plot(q.real)
ax.plot(X.real)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Re{Fourier Transform} [C s]')
ax.set_title('The real part of canned routine is equiv to real part of directly computed  ')
ax = fig.add_subplot(2,1,2)
ax.plot(k, q.imag)
ax.plot(k, X.imag)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Im{Fourier Transform} [C s]')
ax.set_title('The imag part of canned routine is equiv to imag part of directly computed  ')
````

This demonstrates how to handle what the np.fft.fft canned routine returns in order to se it properly.

**2** Now, we can go ahead and use `np.fft.fft`.  Estimate the power spectrum for the Deep Cove time series (again go ahead and skip over gaps, though some of the bigs ones will screw up your frequencies a bit)

  - Use *all* the data.  
  - Report your results in proper SI units.
  - Indicate on your plot at least a couple of important frequencies
  - Comment on the features you see.

````python
dc = dc_raw[np.isfinite(dc_raw)]
N = len(dc)
T = 3600*N
k = np.arange(0,N/2+1.)/T

FT = 3600*np.fft.fft(dc)
FT = FT[:len(FT)/2+1]
PSD = 2./T*np.conjugate(FT)*FT
PSD[0] = PSD[0]/2. #DC signal of Gxx = Sxx
PSD = np.real(PSD)

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
ax.loglog(k,PSD)
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('G_xx [C^2 s]')
ax.set_title('G_xx for the 3 year data of Deep Cove Temperature')
ax.plot((1./(365*24*3600), 1./(365*24*3600)), (1e-9, 10**11), '--', label = 'Seasonal Frequency = 1/(365*24*3600)')
ax.plot((1./(24*3600), 1./(24*3600)), (1e-9, 10**11), '--', label = 'Daily Frequency = 1/(365*24*3600)')
ax.legend(loc='best')
ax.set_xlim(0,1e-4)
````

To validate the computed G_xx, the frequencies corresponding to years and to days are computed and laid against the plot. We see that they superimpose 2 peaks in the PSD. There are several other peaks, perhaps most notably the one located at ~2.4e-5 Hz, or occuring every 12 hours. Is this data describing water temperature? If so, such periodicity might be due to tidal flows that bring in different-temperature water with 12 hour periodicity.

## Q2: Emperically derive distribution of spectral estimate

**1** Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization.  The number of data points $N$ can be relatively small (say 1000).

````python
def Red_Noise(N):
    r = np.random.randn(N)
    red = np.cumsum(r)
    return red

red = Red_Noise(1000)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.plot(red)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.set_title('Red Noise')
````

**2** Write a second function to return the estimate of the power spectrum of $x$.  For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate.  Plot the resulting spectrum.

````python
def Gxx_Calc(sig, sf):
    N = len(sig)
    T = 1./sf*N
    k = np.arange(0,N/2.+1.)/T

    FT = 1./sf*np.fft.fft(sig)
    FT = FT[:len(FT)/2+1]
    PSD = 2./T*np.conjugate(FT)*FT
    PSD[0] = PSD[0]/2. #DC signal of Gxx = Sxx
    PSD = np.real(PSD)
    return PSD, k

q = Gxx_Calc(red,1)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.loglog(q[1],q[0])
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Spectrum [V^2 Hz^-1]')
ax.set_title('Red Noise')
ax.set_xlim(0,0.5)
````

````python
n = 100
N = 1000
Gxxs = np.zeros((N/2.+1,n+1))

for i in range(0,n):
    red = Red_Noise(N)
    Gxxs[:,i] = Gxx_Calc(red,1)[0]   
Gxxs[:,n] = Gxx_Calc(red,1)[1]
    
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.loglog(Gxxs[:,n],Gxxs[:,:n], ',')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Power Spectrum [V^2 Hz^-1]')
ax.set_title('Red Noise Power Spectra for 1000 realizations')
ax.set_xlim(0,0.5)
````

**5** Show that the ratio of the lower and upper bounds of your spectral estimates spread compares favourably with the ratio between the lower and upper 95% confidence bounds of the  $\chi^2_2$ distribution.  Comment on any difference.

````python
inter = stats.chi2.interval(0.95,df=2)

r_chi = (2.*Gxx_m[1:]/inter[1])/(2.*Gxx_m[1:]/inter[0])
r_95 = lo[1:]/hi[1:]

r_95_bin = np.zeros(10)
for i in range (0,10):
    r_95_bin[i] = np.mean(r_95[50*i:50*i+50])
f_bin = np.arange(25,500,50)*0.001

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.plot(f_bin, r_95_bin, label = '95% data binned')
ax.plot(Gxxs[1:,n], r_95, 'b.',label = '95% data')
ax.plot(Gxxs[1:,n], r_chi, label = 'Chi^2 distribution')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Ratio between upper and lower')
ax.set_title('The ratio of limits between Chi^2 and 95% data encompassment')
ax.set_xlim(0,0.5)
ax.legend(loc='best')
````

By binning the data, we see that the ratio of the boundaries for 95% data encompassment is similar to the ratio of the upper and lower error bars for the xi^2 distribution. However, there is a noticable underestimation, which is probably prominantly due to the low number of realizations which cause an unsmooth 95% lower and upper bound. It is probably also due to the small sample size, although to a lesser extent.
