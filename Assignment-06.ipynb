### Error bounds on coherence estimates

<div style='background:#F0F0F0'>  The 95% confidence intervals for the coherence squared $\gamma^2_{xy}(f)$ between two variables $x(t)$ and $y(t)$ are approximately:
$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.  

You **may** use `matplotlib.mlab.csd` for the following question, but you had best use it correctly!
</div>

<div style='background:#F0F0F0'>**1** Make an input process $x(t)$ as random Gaussian noise, and the output process the sum of a noise source and a running average of the input: $y(t)=n(t)+\frac{1}{b}\int_{t-b}^t x(t')dt'$ and plot (Zoom so we can see whats going on, and label your plot).  Make sure you generate a large time series.</div> 

```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy.ma as ma
%matplotlib inline

dt = 1.
b = 10
N = 40000+b

def signals(N,b,dt):
    t = np.arange(0.,N,dt)
    x = np.random.randn(N)
    y = np.zeros(N-b)
    #trapezoidal integration from t_{i-b} to t_{i}
    for i in range(0,len(t)-b):
        y[i] = np.trapz(x[i:i+b+1])
    y =  y/b + 0.1*np.random.randn(N-b)
    return x, y, t

x, y, t = signals(N, b, dt)

fig = plt.figure(figsize=(10,12))
ax = fig.add_subplot(2,1,1)
ax.plot(t,x,'b',label='Input Signal x(t)')
ax.plot(t[b:],y,'r',label='Output Signal y(t)')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.set_title('Input and output signals')
ax.set_xlim(0,np.max(t))
ax.legend(loc = 'best')

ax = fig.add_subplot(2,1,2)
ax.plot(t,x,'b',label='Input Signal x(t)')
ax.plot(t[b:],y,'r',label='Output Signal y(t)')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal [V]')
ax.set_title('Input and output signals')
ax.set_xlim(4000,4400)
ax.legend(loc = 'best')
```

<div style='background:#F0F0F0'>**2** Compute the coherence squared of the input and the output.  Make a plot with the spectra in the upper panel and the coherence squared in the middle panel and the phase in the bottom.

You will need to choose a value for your fft length.  Choose a relative low one to get a smooth plot.
</div>

```python
#I smooth the data slightly with the pad_to attribute

def cohere(x,y,NFFT):
    Gxx,fxx = mlab.psd(x[b:],NFFT=NFFT,Fs=1./dt,window=mlab.window_hanning,
                       noverlap=NFFT/2.)
    Gyy,fyy = mlab.psd(y,NFFT=NFFT,Fs=1./dt,window=mlab.window_hanning,
                       noverlap=NFFT/2.)
    Gxy,fxy = mlab.csd(x[b:],y,NFFT=NFFT,Fs=1./dt,window=mlab.window_hanning,
                       noverlap=NFFT/2.)
    
    gsq = np.absolute(Gxy)**2/Gxx/Gyy
    
    gsq_hi = gsq + np.sqrt(2)*(1+gsq)/(np.sqrt(gsq)*np.sqrt((N-b)/NFFT))
    gsq_lo = gsq - np.sqrt(2)*(1+gsq)/(np.sqrt(gsq)*np.sqrt((N-b)/NFFT))
    
    theta = np.arctan2(np.imag(Gxy),np.real(Gxy))
    
    return Gxx, Gyy, gsq, gsq_hi, gsq_lo, theta, fxx

Gxx, Gyy, gsq, gsq_hi, gsq_lo, theta, f = cohere(x,y,512)
fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(3,1,1)
ax.loglog(f,Gxx,'b',label='Gxx(f)')
ax.plot(f,Gyy,'r',label='Gyy(f)')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Signal [V^2 Hz^-1]')
ax.set_title('Spectra of x(t) and y(t)')
ax.legend(loc = 'best')

ax = fig.add_subplot(3,1,2)
ax.semilogx(f,gsq,'b',label='gamma_sq(f)')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Coherence Squared')
ax.set_title('Coherence Squared between x(t) and y(t)')
ax.legend(loc = 'best')

ax = fig.add_subplot(3,1,3)
ax.semilogx(f,theta,'b',label='phi(Gxy)')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('theta [rads]')
ax.set_title('Phase of cross spectra')
ax.legend(loc = 'best')
```

The coherence signal starts at 1 and then decays. It is also modulated by a periodic function that has a periodicity of 0.1 in the frequency scale. This phenomenon can also be observed in the theta plot, where the cross spectra phase goes in and out of phase with the same periodicity. What's also neat about the phase plot is that it is sensitive to changes in the coherence squared; for example, the phase begins noticably decreasing well before the coherence squared first drops.

</div>

```python
fig = plt.figure(figsize=(10,20))

Gxx, Gyy, gsq, gsq_hi, gsq_lo, theta, f = cohere(x,y,512)
ax = fig.add_subplot(4,1,1)
ax.semilogx(f,gsq,'b',label='NFFT = 512')
ax.fill_between(f,gsq_lo,gsq_hi,alpha=0.5,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Coherence Squared')
ax.set_title('Coherence Squared between x(t) and y(t)')
ax.legend(loc = 'best')
ax.set_xlim(1e-2,0.5)
ax.set_ylim(-1,2)

Gxx, Gyy, gsq, gsq_hi, gsq_lo, theta, f = cohere(x,y,256)
ax = fig.add_subplot(4,1,2)
ax.fill_between(f,gsq_lo,gsq_hi,alpha=0.5,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
ax.semilogx(f,gsq,'b',label='NFFT = 256')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Coherence Squared')
ax.set_title('Coherence Squared between x(t) and y(t)')
ax.legend(loc = 'best')
ax.set_ylim(-1,2)
ax.set_xlim(1e-2,0.5)

Gxx, Gyy, gsq, gsq_hi, gsq_lo, theta, f = cohere(x,y,128)
ax = fig.add_subplot(4,1,3)
ax.fill_between(f,gsq_lo,gsq_hi,alpha=0.5,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
ax.semilogx(f,gsq,'b',label='NFFT = 128')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Coherence Squared')
ax.set_title('Coherence Squared between x(t) and y(t)')
ax.legend(loc = 'best')
ax.set_ylim(-1,2)
ax.set_xlim(1e-2,0.5)

Gxx, Gyy, gsq, gsq_hi, gsq_lo, theta, f = cohere(x,y,64)
ax = fig.add_subplot(4,1,4)
ax.fill_between(f,gsq_lo,gsq_hi,alpha=0.5,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
ax.semilogx(f,gsq,'b',label='NFFT = 64')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Coherence Squared')
ax.set_title('Coherence Squared between x(t) and y(t)')
ax.legend(loc = 'best')
ax.set_ylim(-1,2)
ax.set_xlim(1e-2,0.5)
```

As NFFT decreases, the number of blocks increases, so we see that the error bars tighten. This is due to the error bars being inversely proportional to the inverse square root of block number. Another important note is that the upper and lower bounds diverge past 0 and 1, which are the bounds of the coherence squared. Points where this occurs indicates that NFFT is too high and thus we cannot conclude the value of the coherence squared with any certainty  around these points.

## Q2: Shear turbulence data

```python
time=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/time.txt')
ac=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/ac.txt')
shear=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/s.txt')
```

<div style='background:#F0F0F0'>Download the data abaove. `time` is a measurement time in seconds. `shear` is a little less than 300 s of data from a turbulence shear probe. Units are $s^{-1}$. `ac` is a suitably scaled acceleration of the vehicle containing the shear probes.

**1** Plot the shear time series and the accelerometer time series.  Comment on any similarities and differences.  Comment on the stationarity of the time series.</div>

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(2,1,1)
ax.plot(time,shear, label='Shear')
ax.plot(time,ac, label='Probe Acceleration')
ax.set_xlabel('Time [s]')
ax.set_ylabel('Value')
ax.legend()
```

During times of high-variance shear, the acceleration is also seen to share a similiar 'boost' in variance, for example at t = [200,275]s. The acceleration carries a bit of a 'random walk' whereas the shear is highly stationary.

<div style='background:#F0F0F0'> **2** Compute the power spectra of the two signals and the coherence squared (you could do the phase too, but it is not too thrilling).  Comment on the spectral similarities and the coherence.</div>

```python
def cohere(x,y,NFFT,dt):
    Gxx,fxx = mlab.psd(x,NFFT=NFFT,Fs=1./dt,window=mlab.window_hanning,
                       noverlap=NFFT/2.)
    Gyy,fyy = mlab.psd(y,NFFT=NFFT,Fs=1./dt,window=mlab.window_hanning,
                       noverlap=NFFT/2.)
    Gxy,fxy = mlab.csd(x,y,NFFT=NFFT,Fs=1./dt,window=mlab.window_hanning,
                       noverlap=NFFT/2.)
    gsq = np.absolute(Gxy)**2/Gxx/Gyy
    return Gxx, Gyy, Gxy, gsq, fxx

dt = 300./len(ac)

Gaa, Gss, Gas, gsq, fss = cohere(ac, shear, 512, dt)

fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(2,1,1)
ax.loglog(fss,Gss, label='Gss')
ax.loglog(fss,Gaa, label='Gaa')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Gss')
ax.legend(loc='best')

ax = fig.add_subplot(2,1,2)
ax.loglog(fss,gsq,label='gamma^2')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('gamma^2')
ax.legend(loc='best')
```

Its clear that there are 2 primary peaks in the acceleration of the vehicle that are influencing the shear data. This can be seen in all three plots at around 17Hz and 40Hz.

<div style='background:#F0F0F0'> **3** Compute a spectra that has the contamination from the vibration removed using the linear technique described in the notes. Comment on how well your correction worked.</div>

```python
Gnn =  Gss-abs(Gas)**2/Gaa

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,1,1)
ax.loglog(fss,Gss, label='Gss')
ax.loglog(fss,Gaa, label='Gaa')
ax.loglog(fss,Gnn,label='Gnn')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Gss')
ax.legend(loc='best')
```

After the filtering, the peak at 17Hz is fully removed and the peak at 40Hz is fractionally removed. Otherwise, the shear data is practically untouched, so the filtering has been partially successful.

<div style='background:#F0F0F0'>**4** Chose two separate periods of the time series by eye, one that is "turbulent" and one that is "quiet", and plot the raw and corrected shear spectra.  Comment on the differences and the effect of the correction on each spectral estimate.    </div>

```python
Gaa_t, Gss_t, Gas_t, gsq_t, fss_t = cohere(ac[29600:37037], shear[29600:37037], 512, dt)
Gnn_t =  Gss_t-abs(Gas_t)**2/Gaa_t
Gaa_q, Gss_q, Gas_q, gsq_q, fss_q = cohere(ac[11111:18500], shear[11111:18500], 512, dt)
Gnn_q =  Gss_q-abs(Gas_q)**2/Gaa_q

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,1,1)
ax.loglog(fss_t,Gss_t, label='Gss')
ax.loglog(fss_t,Gnn_t,label='Gnn')
ax.set_title('Turbulent Data')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Gss')
ax.legend(loc='best')

ax = fig.add_subplot(2,1,2)
ax.loglog(fss_q,Gss_q, label='Gss')
ax.loglog(fss_q,Gnn_q,label='Gnn')
ax.set_title('Quiet Data')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('Gss')
ax.legend(loc='best')




```

In the turbulent data, the filtering does an excellent job of preserving the data that isn't affected by the primary frequency disturbances brought on by the vehicle's acceleration. Filtering on the quiet data seems to have the effect of stripping some of the values in the low frequency, which probably has to do with the fact that the acceleration is calmer and thus has lower frequency components.
