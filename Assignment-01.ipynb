Consider a coin that is twice as likely to come up heads as tails. To put a number on it, call tails $x=-1/2$ and heads $x=+1/2$.

**3** Write a function using the `np.random.norm` function to simulate the coin flip.  It shoudl return $+1/2$ twice as often as $-1/2$.  Demonstrate that your function works.
<hr>

```python
import scipy.stats as stat

def getrolls(N,ratio=False,stat=False):
    
    values = np.floor(3*np.random.rand(1,N))
    H_counter = 0.
    T_counter = 0.
    
    for i in range(0,N):
        if values[0,i] == 0.:
            values[0,i] = -1./2
            T_counter += 1
        else:
            values[0,i] = 1./2
            H_counter += 1
    
    if stat:
        return np.mean(values)
    if ratio:
        print "Ratio of heads to tails is", float(H_counter/T_counter)

getrolls(10000,ratio=True,stat=False)
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{roll}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll.  
<hr>

```python
import matplotlib.mlab as mlab

def experiments(Nexp, N):
    mean_array = np.zeros(Nexp)
    for i in range (0,Nexp):
        mean_array[i] = getrolls(N,ratio=False,stat=True)
    return mean_array, np.mean(mean_array), np.std(mean_array)
    
Nexp = 2000
rolls_2 = experiments(Nexp,2)
rolls_16 = experiments(Nexp,16)
rolls_64 = experiments(Nexp,64)
rolls_256 = experiments(Nexp,256)


#Note, histograms have been normalized to compare to theoretical normal distribution pdfs. 
#The convergence is better and better for higher # of flips in each experiment

fig=plt.figure(figsize=(8,20))
x = np.arange(-0.5,0.5,0.01) #x range for theoretical curves

ax=fig.add_subplot(4,1,1)
ax.plot(x,mlab.normpdf(x,rolls_2[1],rolls_2[2]))
ax.hist(rolls_2[0],normed=True)
ax.set_xlim(-0.5,0.5)
ax.set_xlabel('Value')
ax.set_ylabel('Occurences')
ax.set_title('2 flip experiment (2000 runs)')

ax=fig.add_subplot(4,1,2)
ax.plot(x,mlab.normpdf(x,rolls_16[1],rolls_16[2]))
ax.hist(rolls_16[0],normed=True)
ax.set_xlim(-0.2,0.55)
ax.set_xlabel('Value')
ax.set_ylabel('Occurences')
ax.set_title('16 flip experiment (2000 runs)')

ax=fig.add_subplot(4,1,3)
ax.plot(x,mlab.normpdf(x,rolls_64[1],rolls_64[2]))
ax.hist(rolls_64[0],normed=True)
ax.set_xlim(-0.05,0.4)
ax.set_xlabel('Value')
ax.set_ylabel('Occurences')
ax.set_title('64 flip experiment (2000 runs)')

ax=fig.add_subplot(4,1,4)
ax.plot(x,mlab.normpdf(x,rolls_256[1],rolls_256[2]))
ax.hist(rolls_256[0],normed=True)
ax.set_xlim(0.05,0.3)
ax.set_xlabel('Value')
ax.set_ylabel('Occurences')
ax.set_title('256 flip experiment (2000 runs)')
```

As you can see, the histograms begin approaching the corresponding normal distribution as N goes to infinity. The difference between the convergence of the coin flip and the dice roll is most apparent in the low number flips. In the dice roll, there are are a multitude of possible means even if each experiment contains only 2 measurements. But in the coin flip experiment, 2 measurements can only yield 3 different outcomes in the mean, as shown in the first histogram. This limits the convergence of the sample mean to the mean.

**5** Replot the histograms, removing the mean and normalizing by the standard deviation.  Also show the CDF for normal distribution with mean zero and standard deviation of 1.
<hr>

```python
fig=plt.figure(figsize=(8,5))

# rolls_#[1] is the sample mean of the means, and rolls_#[2] is the sample std of the means
ax=fig.add_subplot(1,1,1)
ax.hist((rolls_2[0]-rolls_2[1])/rolls_2[2],20,label='2 flips')
ax.hist((rolls_16[0]-rolls_16[1])/rolls_16[2],20,label='16 flips')
ax.hist((rolls_64[0]-rolls_64[1])/rolls_64[2],20,label='64 flips')
ax.hist((rolls_256[0]-rolls_256[1])/rolls_256[2],20,label='256 flips')
ax.set_xlabel('Value')
ax.set_ylabel('Occurences')
ax.set_title('Normalizing by standard deviation makes the histograms have equal widths')
ax.legend()

dx = 0.01
x = np.arange(-3.,3.,dx) #x range for theoretical curves
pdf = mlab.normpdf(x,0,1)
cdf = np.cumsum(pdf*dx)
fig=plt.figure(figsize=(8,5))
ax=fig.add_subplot(1,1,1)
ax.plot(x,cdf)
ax.set_xlabel('Value')
ax.set_ylabel('Cumulative probability')
ax.set_title('The CDF of a normal distribution is Sigmoidal')
```
