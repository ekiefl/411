**3** Write a function using the `np.random.norm` function to simulate the coin flip.  It shoudl return $+1/2$ twice as often as $-1/2$.  Demonstrate that your function works.
<hr>

```python
import scipy.stats as stat

def getrolls(N,ratio=False,stat=False):
    
    values = np.floor(3*np.random.rand(1,N))
    H_counter = 0.
    T_counter = 0.
    
    for i in range(0,N):
        if values[0,i] == 0.:
            values[0,i] = -1./2
            T_counter += 1
        else:
            values[0,i] = 1./2
            H_counter += 1
    
    if stat:
        return np.mean(values)
    if ratio:
        print float(H_counter/T_counter)

getrolls(10000,ratio=True,stat=False)
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{roll}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll.  
<hr>

```python
import matplotlib.mlab as mlab

def experiments(Nexp, N):
    mean_array = np.zeros(Nexp)
    for i in range (0,Nexp):
        mean_array[i] = getrolls(N,ratio=False,stat=True)
    return mean_array, np.mean(mean_array), np.std(mean_array)
    
Nexp = 2000    
rolls_2 = experiments(Nexp,2)
rolls_16 = experiments(Nexp,16)
rolls_64 = experiments(Nexp,64)
rolls_256 = experiments(Nexp,256)

fig=plt.figure(figsize=(8,5))
ax=fig.add_subplot(1,1,1)
ax.hist(rolls_2[0],20)
ax.hist(rolls_16[0],20)
ax.hist(rolls_64[0],20)
ax.hist(rolls_256[0],20)


#trying to superimpose theoretical curves
x = np.arange(-0.5,0.5,0.01)
fig=plt.figure(figsize=(8,20))
ax=fig.add_subplot(4,1,1)
ax.plot(mlab.normpdf(x,rolls_256[1],rolls_256[2]))
print rolls_256[1],rolls_256[2]
ax.hist(rolls_256[0],normed=1)

```
